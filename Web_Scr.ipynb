```python
import cloudscraper
from bs4 import BeautifulSoup
import csv
import time

# CREATE SCRAPER
scraper = cloudscraper.create_scraper()

# CSV FILE
filecsv = open('PC_Product.csv', 'w', newline='', encoding='utf8')

csv_columns = ['Brand', 'model', 'price', 'availability', 'img']

writer = csv.DictWriter(filecsv, fieldnames=csv_columns)
writer.writeheader()

# URL
url = 'https://kimostore.net/collections/pc-components?page='

# LOOP PAGES
for page in range(1, 15):

    print(f"---------------- PAGE {page} ----------------")

    time.sleep(3)

    response = scraper.get(url + str(page))
    soup = BeautifulSoup(response.text, 'html.parser')

    # PRODUCTS
    products = soup.find_all('div', class_='product-item')

    for product in products:

        # IMAGE
        try:
            img_tag = product.find('img')

            img = img_tag.get('data-src') or img_tag.get('src')

            if img.startswith('//'):
                img = 'https:' + img

        except:
            img = "Not Available"

        # PRICE
        try:
            price_tag = product.find('span', class_='price')

            price = (
                price_tag.text
                .strip()
                .replace('Sale price', '')
                .replace('EGP', '')
                .replace(',', '')
                .strip()
            )

        except:
            price = "Not Available"

        # MODEL
        try:
            model = product.find(
                'a',
                class_='product-item__title'
            ).text.strip()

        except:
            model = "Not Available"

        # BRAND
        try:
            brand = product.find(
                'a',
                class_='product-item__vendor'
            ).text.strip()

        except:
            brand = "Not Available"

        # AVAILABILITY
        try:
            availability = product.find(
                'span',
                class_='inventory'
            ).text.strip()

        except:
            availability = "Not Available"

        print(brand, model, price)

        # SAVE CSV
        writer.writerow({
            'Brand': brand,
            'model': model,
            'price': price,
            'availability': availability,
            'img': img
        })

filecsv.close()

```
